**ARIMA Model for testing the time series data **

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('bitcoin_with_metrics.csv')
```

```python
df
```

```python
#setting datetime to index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
#target variables
target_variable_high = '%_Diff_From_High_Next_5_Days'
target_variable_low = '%_Diff_From_Low_Next_5_Days'

# Split data into training and testing sets
train_size = int(len(df) * 0.8)  # 80% for training
train_high, test_high = df[target_variable_high][:train_size], df[target_variable_high][train_size:]
train_low, test_low = df[target_variable_low][:train_size], df[target_variable_low][train_size:]
```

```python
# ARIMA for %_Diff_From_High
model_high = ARIMA(train_high.dropna(), order=(5, 1, 0))
model_high_fit = model_high.fit()
forecast_high = model_high_fit.forecast(steps=len(test_high.dropna()))  # Forecast for the test set, ensuring the same length as the test set after removing NaNs

# ARIMA for %_Diff_From_Low
model_low = ARIMA(train_low.dropna(), order=(5, 1, 0))  # Adjust order as necessary
model_low_fit = model_low.fit()
forecast_low = model_low_fit.forecast(steps=len(test_low.dropna()))  # Forecast for the test set, ensuring the same length as the test set after removing NaNs
```

```python
# Evaluate the model
rmse_high = np.sqrt(mean_squared_error(test_high.dropna(), forecast_high))
rmse_low = np.sqrt(mean_squared_error(test_low.dropna(), forecast_low))

print(f'ARIMA RMSE for High: {rmse_high}')
print(f'ARIMA RMSE for Low: {rmse_low}')
```

```python
# Visualization
plt.figure(figsize=(14, 7))
plt.plot(test_high.dropna().index, test_high.dropna(), label='Actual % Diff From High', color='blue') # Align the index for plotting after removing NaNs
plt.plot(test_high.dropna().index, forecast_high, label='Forecast % Diff From High', color='orange') # Align the index for plotting after removing NaNs
plt.title('% Diff From High Forecast')
plt.xlabel('Date')
plt.ylabel('% Diff From High')
plt.legend()
plt.show()

plt.figure(figsize=(14, 7))
plt.plot(test_low.dropna().index, test_low.dropna(), label='Actual % Diff From Low', color='green') # Align the index for plotting after removing NaNs
plt.plot(test_low.dropna().index, forecast_low, label='Forecast % Diff From Low', color='red') # Align the index for plotting after removing NaNs
plt.title('% Diff From Low Forecast')
plt.xlabel('Date')
plt.ylabel('% Diff From Low')
plt.legend()
plt.show()
```

LSTM

```python

from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# setting date as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
# Prepare target variables
target_variable_high = '%_Diff_From_High_Next_5_Days'
target_variable_low = '%_Diff_From_Low_Next_5_Days'

# Extract features - Corrected column names to match DataFrame
features = df[['Days_Since_High_Last_7_Days', '%_Diff_From_High_Last_7_Days',
                'Days_Since_Low_Last_7_Days', '%_Diff_From_Low_Last_7_Days']]

# Add target variables
features[target_variable_high] = df[target_variable_high]
features[target_variable_low] = df[target_variable_low]

# Drop rows with missing values
features.dropna(inplace=True)

# Split the data into features (X) and targets (y)
X = features.iloc[:, :-2].values  # All columns except targets
y_high = features[target_variable_high].values  # Target variable high
y_low = features[target_variable_low].values  # Target variable low
```

```python
# Standard scale the features
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)

# Scale the target variables
scaler_y_high = StandardScaler()
y_high_scaled = scaler_y_high.fit_transform(y_high.reshape(-1, 1))

scaler_y_low = StandardScaler()
y_low_scaled = scaler_y_low.fit_transform(y_low.reshape(-1, 1))
```

```python
# Split the data into training and testing sets
train_size = int(len(X_scaled) * 0.8)
X_train, X_test = X_scaled[:train_size], X_scaled[train_size:]
y_train_high, y_test_high = y_high_scaled[:train_size], y_high_scaled[train_size:]
y_train_low, y_test_low = y_low_scaled[:train_size], y_low_scaled[train_size:]
```

```python
# Reshape input to be [samples, time steps, features]
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))
```

```python
# Build and train LSTM model for High
model_high = Sequential()
model_high.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model_high.add(Dropout(0.2))
model_high.add(LSTM(units=50))
model_high.add(Dropout(0.2))
model_high.add(Dense(1))
model_high.compile(loss='mean_squared_error', optimizer='adam')

# Fit the model
history_high = model_high.fit(X_train, y_train_high, epochs=100, batch_size=32, validation_data=(X_test, y_test_high))
```

```python
# Make predictions for High
predictions_high = model_high.predict(X_test)
predictions_high = scaler_y_high.inverse_transform(predictions_high)

# Build and train LSTM model for Low
model_low = Sequential()
model_low.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model_low.add(Dropout(0.2))
model_low.add(LSTM(units=50))
model_low.add(Dropout(0.2))
model_low.add(Dense(1))
model_low.compile(loss='mean_squared_error', optimizer='adam')

# Fit the model for Low
history_low = model_low.fit(X_train, y_train_low, epochs=100, batch_size=32, validation_data=(X_test, y_test_low))

# Make predictions for Low
predictions_low = model_low.predict(X_test)
predictions_low = scaler_y_low.inverse_transform(predictions_low)
```

```python
# Calculate RMSE
rmse_high = sqrt(mean_squared_error(y_test_high, predictions_high))
rmse_low = sqrt(mean_squared_error(y_test_low, predictions_low))
print(f"RMSE for High Predictions: {rmse_high}")
print(f"RMSE for Low Predictions: {rmse_low}")

# Plot predictions for High
plt.figure(figsize=(14, 7))
plt.plot(y_test_high, label='Actual High % Diff', color='blue')
plt.plot(predictions_high, label='Predicted High % Diff', color='red')
plt.title('High % Diff Predictions vs Actual')
plt.xlabel('Time')
plt.ylabel('High % Diff')
plt.legend()
plt.show()

# Plot predictions for Low
plt.figure(figsize=(14, 7))
plt.plot(y_test_low, label='Actual Low % Diff', color='blue')
plt.plot(predictions_low, label='Predicted Low % Diff', color='red')
plt.title('Low % Diff Predictions vs Actual')
plt.xlabel('Time')
plt.ylabel('Low % Diff')
plt.legend()
plt.show()
```

RandomForest Regressor Model

```python

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV

```

```python
df.dropna(inplace=True) # Changed to inplace=True to modify df directly

# Define features and target variable
features = [f'Days_Since_High_Last_7_Days', f'%_Diff_From_High_Last_7_Days',
            f'Days_Since_Low_Last_7_Days', f'%_Diff_From_Low_Last_7_Days']
target_high = f'%_Diff_From_High_Next_5_Days'
target_low = f'%_Diff_From_Low_Next_5_Days'
```

```python
# Splitting the data
X = df[features]
y_high = df[target_high]
y_low = df[target_low]

# Standard Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train_high, y_test_high = train_test_split(X_scaled, y_high, test_size=0.2, random_state=42)
_, _, y_train_low, y_test_low = train_test_split(X_scaled, y_low, test_size=0.2, random_state=42)
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None] # Changed 'auto' to None
}

# Initialize RandomForestRegressor
rf = RandomForestRegressor(random_state=42)
```

```python
# Grid Search
grid_search_high = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search_high.fit(X_train, y_train_high)
best_rf_high = grid_search_high.best_estimator_

# Make predictions and evaluate for High
y_pred_high = best_rf_high.predict(X_test)
rmse_high = np.sqrt(mean_squared_error(y_test_high, y_pred_high))


print(f"Best Parameters for High Prediction: {grid_search_high.best_params_}")
print(f"Random Forest RMSE for High Prediction: {rmse_high}")
```

```python
# Repeat for Low
grid_search_low = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search_low.fit(X_train, y_train_low)
best_rf_low = grid_search_low.best_estimator_

# Make predictions and evaluate for Low
y_pred_low = best_rf_low.predict(X_test)
rmse_low = np.sqrt(mean_squared_error(y_test_low, y_pred_low))
```

```python
print(f"Best Parameters for Low Prediction: {grid_search_low.best_params_}")
print(f"Random Forest RMSE for Low Prediction: {rmse_low}")
```

```python
import matplotlib.pyplot as plt

# Plot for High Prediction
plt.figure(figsize=(14, 5))
plt.plot(y_test_high.values, label='Actual High')
plt.plot(y_pred_high, label='Predicted High', linestyle='--')
plt.title('Actual vs Predicted for % Diff From High')
plt.xlabel('Samples')
plt.ylabel(f'% Diff From High Next 7 Days')
plt.legend()
plt.show()

# Plot for Low Prediction
plt.figure(figsize=(14, 5))
plt.plot(y_test_low.values, label='Actual Low')
plt.plot(y_pred_low, label='Predicted Low', linestyle='--')
plt.title('Actual vs Predicted for % Diff From Low')
plt.xlabel('Samples')
plt.ylabel(f'% Diff From Low Next 7 Days')
plt.legend()
plt.show()
```
